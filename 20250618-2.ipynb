#### 금천구 대상으로 각 주유소별 휘발유, 국제 유가(WTI,브렌트,두바이유) 상관계수

```python
import pandas as pd 
import numpy 
import random
```

```python
oil_1 = pd.read_csv("./주유소/상반기 주유소 판매가격.csv" , encoding='cp949')
oil_2 = pd.read_csv("./주유소/하반기 주유소 판매가격.csv" , encoding='cp949')
```

```python
oil=pd.concat([oil_1,oil_2],ignore_index=True)
```

```python
oil
```

```python
location=oil['지역'].unique()
location

brand= oil['상표'].unique()
brand
```

```python
oil_adress=oil.drop_duplicates(['주소'])
oil_adress
```

```python
list=[]
count={}
for i in location:
    df=oil_adress[oil_adress['지역']==i]
    for j in brand:
        count_brand={j:len(df.loc[df['상표']==j])}
        list.append(count_brand)
    count[i]=list
    list=[]
count
```

```python
df = pd.DataFrame.from_dict(count)
df

파이팅해라 브레드킹한핑퐁
```
