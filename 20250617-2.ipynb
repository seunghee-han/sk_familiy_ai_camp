```python
import pandas as pd 
import numpy 
import random
```

```python
train = pd.read_csv("./data/train.csv")
test = pd.read_csv("./data/test.csv")
```

```python
train.isnull().sum()
```

```python
train_name_df = train['Name'].str.split("[,.]", expand=True)
```

```python
train_name_df.head(1)
```

```python
train_name_df.columns = ['family_name', 'honorific', 'name', 'dummy']
```

```python
train_name_df.drop('dummy', axis=1, inplace=True)
```

```python
train_name_df['honorific'].value_counts()
```

```python
train = pd.concat([train, train_name_df], axis=1)
```

```python
df = pd.merge(train, train.groupby('honorific', as_index=False)[['Age']].mean(), 
         left_on=['honorific'], right_on=['honorific'], how='left')
```

```python
df.loc[df.Age_x.isnull(), 'Age_x'] = df['Age_y']
```
